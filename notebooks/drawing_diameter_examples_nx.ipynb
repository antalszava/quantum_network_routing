```python
import matplotlib2tikz
from networkx import grid_graph
import math
import matplotlib.pyplot as plt
G = nx.Graph()
for x in list(itertools.product(range(10), repeat=2)):
    v1 = x[0]
    v2 = x[1]
    if 1==abs(v1-v2):
        G.add_edge(v1,v2)

        
avg = nx.average_shortest_path_length(G)
diam = nx.diameter(G)
print(avg)
print(diam)
print(diam/avg)
nx.draw_networkx(G)
```
