# Run global knowledge for maxdth=16

```python
samples = 1000
max_th = 2
global_knowledge_results = []
start = time.time()
factory = graph_edge_factory.GraphEdgesFactory(original_capacity = 0)
graph_edges = factory.generate_deterministic_graph_edges(factory.deterministic_link)
arguments = {'algorithm': routing_algorithms.global_knowledge_init, 'graph_edges': graph_edges,
             'link_prediction': False, 'exponential_scale': True}
local_result, length = routing_simulation.run_algorithm_for_graphs(50, samples, arguments)
global_knowledge_results.append(local_result)
th = 0
for th in range(max_th + 1):
    threshold = 2 ** th
    factory = graph_edge_factory.GraphEdgesFactory(distance_threshold=threshold, max_threshold=2 ** max_th)
    graph_edges = factory.generate_deterministic_graph_edges(factory.deterministic_link)
    arguments = {'algorithm': routing_algorithms.global_knowledge_init, 'graph_edges': graph_edges,
                 'link_prediction': False, 'exponential_scale': True}
    local_result, length = routing_simulation.run_algorithm_for_graphs(50, samples, arguments)
    global_knowledge_results.append(local_result)
    end = time.time()
write_results_to_file(global_knowledge_results, 'global_knowledge_det_maxdth_' + str(max_th), '' , (end - start))
```
